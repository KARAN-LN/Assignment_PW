3. Explain the following termsg
 The difference in np.array, np.asarray and np.asanyarrayX
 The difference between Deep copy and shallow copyX

 1. np.array, np.asarray, np.asanyarray
np.array: Always creates a new copy of the input data, even if it's already a NumPy array.
np.asarray: Converts to a NumPy array but doesn't copy if the input is already an array.
np.asanyarray: Similar to np.asarray, but it preserves subclasses of ndarray (e.g., masked arrays).
2. Deep Copy vs Shallow Copy
Shallow Copy: Copies the outer object, but nested objects (like lists or arrays) are shared between the original and the copy.
Example: copy.copy(a)
Deep Copy: Creates a complete independent copy, including all nested objects.
Example: copy.deepcopy(a)

1. Difference Between np.array(), np.asarray(), and np.asanyarray():
np.array(): Always copies the data, even if the input is already a NumPy array.
np.asarray(): Only creates a new array if the input isn't already a NumPy array. It avoids copying if the input is already an array.
np.asanyarray(): Similar to np.asarray(), but preserves the subclass (e.g., np.ma.MaskedArray), while avoiding unnecessary copying.
2. Shallow Copy vs Deep Copy:
Shallow Copy: Copies the outer object, but nested objects are shared. Changes to nested objects affect both the original and copy.
Deep Copy: Copies the entire object and all nested objects. The original and copy are completely independent.

```python
# 5.Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array
#perform the following operations:
# a)Extract all even integers from array.
#b)Extract all odd integers from arrayX
import numpy as np

# Create a 5x6 array with random integers between 1 and 10
arr = np.random.randint(1, 11, (5, 6))

# Extract all even integers from the array
even_integers = arr[arr % 2 == 0]

# Extract all odd integers from the array
odd_integers = arr[arr % 2 != 0]

print("Original Array:\n", arr)
print("\nEven Integers:\n", even_integers)
print("\nOdd Integers:\n", odd_integers)
```

```python
#14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
#Perform the following tasks using Matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create the Time-Series Dataset
# Generate a date range for the past 30 days
date_range = pd.date_range(start="2023-10-01", periods=30, freq='D')

# Generate random temperature values between 20°C and 35°C
temperature = np.random.uniform(20, 35, size=30)

# Generate random humidity values between 40% and 90%
humidity = np.random.uniform(40, 90, size=30)

# Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 2: Plotting with Two Y-Axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the 'Temperature' on the left y-axis (ax1)
ax1.plot(df['Date'], df['Temperature'], color='tab:blue', marker='o', label='Temperature (°C)')
ax1.set_xlabel('Date')  # Label x-axis
ax1.set_ylabel('Temperature (°C)', color='tab:blue')  # Left y-axis label
ax1.tick_params(axis='y', labelcolor='tab:blue')  # Color y-axis ticks for Temperature

# Create a second y-axis (ax2) for the 'Humidity'
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='tab:orange', marker='s', label='Humidity (%)')
ax2.set_ylabel('Humidity (%)', color='tab:orange')  # Right y-axis label
ax2.tick_params(axis='y', labelcolor='tab:orange')  # Color y-axis ticks for Humidity

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()

```

```python
#15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
#tasks using Matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)

plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
#16. Set the title of the plot as 'Histogram with PDF Overlay'.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(0, 1, 1000)

# Plot the histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Get the x-axis limits
xmin, xmax = plt.xlim()

# Create a range of x values for the PDF line
x = np.linspace(xmin, xmax, 100)

# Compute the normal distribution PDF
p = norm.pdf(x, 0, 1)

# Overlay the PDF as a line plot
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
#Kçg Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
#origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

quadrants = []
for xi, yi in zip(x, y):
    if xi > 0 and yi > 0:
        quadrants.append('Quadrant 1')
    elif xi < 0 and yi > 0:
        quadrants.append('Quadrant 2')
    elif xi < 0 and yi < 0:
        quadrants.append('Quadrant 3')
    else:
        quadrants.append('Quadrant 4')

df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Quadrant': quadrants
})

plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', hue='Quadrant', palette='Set1', data=df, s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')
plt.tight_layout()
plt.show()
```

```python
#18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
#Wave Function'8
from bokeh.plotting import figure, show
import numpy as np

# Generate data for sine wave
x = np.linspace(0, 4 * np.pi, 100)  # 100 points from 0 to 4*pi
y = np.sin(x)

# Create a figure
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y', tools="pan,box_zoom,reset")

# Plot the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, line_color="blue")

# Show grid lines
p.grid.grid_line_color = "gray"
p.grid.grid_line_alpha = 0.3

# Display the plot
show(p)
```

```python
#19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
#values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
#Bar Chart'8
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 20, len(categories))  # Random values between 1 and 20

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Set up the Bokeh data source
source = ColumnDataSource(df)

# Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value', tools="pan,box_zoom,reset")

# Plot the bars with color based on value
p.vbar(x='Category', top='Value', source=source, width=0.5,
       color='blue', legend_field='Category', line_color='white')

# Add hover tool to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Display the legend and make it interactive
p.legend.title = 'Category'
p.legend.location = "top_left"

# Display grid lines and adjust their appearance
p.grid.grid_line_color = "gray"
p.grid.grid_line_alpha = 0.3

# Show the plot
output_file("random_categorical_bar_chart.html")
show(p)
```
