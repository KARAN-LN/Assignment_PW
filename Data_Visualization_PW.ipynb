```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility

# Generate random data for 3 features, with correlation between them
n_samples = 1000
x = np.random.normal(0, 1, n_samples)  # Feature 1
y = 2 * x + np.random.normal(0, 0.1, n_samples)  # Feature 2 (correlated with x)
z = -0.5 * x + 0.5 * y + np.random.normal(0, 0.1, n_samples)  # Feature 3 (correlated with x and y)

# Create a DataFrame to hold the features
data = pd.DataFrame({
    'Feature 1': x,
    'Feature 2': y,
    'Feature 3': z
})

# Step 2: Compute the correlation matrix
correlation_matrix = data.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f', linewidths=0.5)

# Add title and labels
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.
import numpy as np
import pandas as pd
import plotly.express as px

# Step 1: Generate the sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create DataFrame
df = pd.DataFrame(data)

# Step 2: Pivot the data to create a matrix with months as rows and days as columns
sales_matrix = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')

# Step 3: Create the heatmap using Plotly
fig = px.imshow(sales_matrix,
                color_continuous_scale='Viridis',
                labels={'color': 'Average Sales'},
                title='Sales Heatmap: Variation Across Months and Days')

# Show the plot
fig.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
import numpy as np

# Step 1: Generate x and y data
x = np.linspace(0, 10, 100)  # x-values from 0 to 10
y = np.sin(x)  # y-values as the sine of x

# Step 2: Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y')

# Step 3: Plot the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Step 4: Show the plot
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show
import numpy as np
import pandas as pd

# Step 1: Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 10  # Random x-values between 0 and 10
y = np.random.rand(n) * 10  # Random y-values between 0 and 10
sizes = np.random.randint(10, 100, size=n)  # Random sizes for the markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # Random colors

# Step 2: Create a Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Step 3: Plot the scatter points
p.scatter(x, y, size=sizes, color=colors, legend_field="color", fill_alpha=0.6)

# Step 4: Show the plot
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable inline plotting in Jupyter Notebook
output_notebook()

# Generate random data (1000 samples from a standard normal distribution)
data_hist = np.random.randn(1000)

# Compute the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Normally Distributed Data",
           x_axis_label='Value', y_axis_label='Frequency',
           tools="pan, box_zoom, reset, hover, save")

# Create the histogram by plotting the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np

# Enable inline plotting in Jupyter Notebook
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)

# Create x and y axis values
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a meshgrid for x and y
xx, yy = np.meshgrid(x, y)

# Flatten the meshgrid to get coordinates for the cells
x_coords = xx.flatten()
y_coords = yy.flatten()
values = data_heatmap.flatten()

# Create a Bokeh figure
p = figure(title="Heatmap",
           x_axis_label='X', y_axis_label='Y',
           tools="hover,pan,box_zoom,reset")

# Create a linear color map based on the values
color_mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(values), high=max(values))

# Plot the heatmap using rect (rectangles)
p.rect(x=x_coords, y=y_coords, width=0.1, height=0.1,
       source={'x': x_coords, 'y': y_coords, 'value': values},
       line_color="white", fill_color=color_mapper)

# Add a color bar for the heatmap
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
